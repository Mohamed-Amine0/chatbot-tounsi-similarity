```python
from chatterbot.trainers import ListTrainer
from chatterbot import ChatBot
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
chatbot = ChatBot('MonChatbot')
trainer = ListTrainer(chatbot)

# entraînez le chatbot sur les données de WhatsApp
conversation = [
    "Bonjour",
    "Salut, comment ça va?",
    "Ça va bien, merci. Et toi?",
    "Je vais bien aussi, merci.",
    "Qu'est-ce que tu fais?",
    "Je discute avec toi. Et toi?",
    "Je travaille sur un projet.",
    "C'est cool! Quel genre de projet?",
    "Un projet de chatbot.",
    "Génial! Tu as besoin d'aide pour cela?",
    "Non, je pense que je suis bon pour le moment. Merci!"
]

trainer.train(conversation)
```

```python
prompt = "Bonjour, comment ça va?"

# créez une liste des entrées de conversation
sentences = list((conversation))
print(sentences)
```

```python
# ajoutez le prompt à la liste des phrases
sentences.append(prompt)
print(sentences)
```

```python
# initialisez un vecteur TfidfVectorizer pour extraire les caractéristiques des phrases
tfidf_vectorizer = TfidfVectorizer()

# créez une matrice Tfidf à partir des phrases
tfidf_matrix = tfidf_vectorizer.fit_transform(sentences)

print(tfidf_matrix)
```

```python
L=tfidf_matrix.todense().tolist()
```

```python
import pandas as pd
df=pd.DataFrame(L,columns=tfidf_vectorizer.get_feature_names())
df
```

```python
# calculez la similarité cosinus entre le prompt et chaque phrase dans la matrice
cosine_similarities = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1])
print(cosine_similarities)
```

```python
# triez les similarités cosinus dans l'ordre décroissant
similarity_list = cosine_similarities.flatten()
print(similarity_list)
```

```python
index = similarity_list.argsort()
print(index)
```

```python

# trouvez l'entrée de conversation la plus similaire
response = conversation[index[-1]+1]

print(response)
```
